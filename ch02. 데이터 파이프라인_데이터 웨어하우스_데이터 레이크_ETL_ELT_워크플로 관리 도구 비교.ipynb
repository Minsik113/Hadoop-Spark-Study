- 다양한 소스에서 데이터를 변환하고 옮기는 일련의 과정을 구성한 시스템

# 3. 데이터 파이프라인 패턴

- 초기에는 종류마다 데이터분석시스템이 따로 존재하였다.
    - 해당 분석 시스템을 통해 분석 결과가 도출되었다.
    - 관리해야하는 포인트도 많아지고 불편함이 존재

> 모든 데이터를 모아서 하나의 분석시스템을 만들고, 각 데이터별 분석 결과를 뽑아보고자 하는 시스템이 생기게 되었다. -> 데이터 웨어하우스

- 분석을 위한 데이터를 데이터 웨어하우스라는 곳으로 모음
- 데이터 웨어하우스는 일반적으로 RDB와는 대량의 데이터를 처리하고 오랫동안 보관하는 것에 최적화 되어 있다.
    - 대량의 데이터를 처리하는데에는 유리하지만, 소량의 데이터를 사용하는데에는 적합하지 않다. 
    - 데이터 웨어하우스를 통해서 통합된 '분석보고서'나 '대시보드'를 운영할 수 있다.
    - 데이터 분석을 위해 데이터 웨어하우스를 많은 회사에서 도입하게 되었다.

## 3-1. 데이터 웨어하우스(Data Warehouse)
- 기존의 DB는 관계형 데이터베이스에 저장을 함.
- 데이터 웨어하우스에 저장할 때, 정규화를 통해서 스키마를 설계하고, 그에 맞게 데이터를 저장한다.
    - 정규화를 하는 이유는 db업데이트 시 최소한의 데이터만 저장이 되도록 하기 위함이다. 작은형태로 만들어 쪼개게 된다.
- RDB에 있는 데이터를 그대로 데이터웨어하우스에 저장하면 분석을 위한 쿼리에는 비효율적이었다.
- 분석을 위한 쿼리를 위해서는 여러 테이블을 조인해야했음. 그래서 테이블을 오른쪽 그림처럼 재구성하여 저장함.


- 데이터 웨어하우스에는 가운데에 펙트 테이블(Fact Table)을 두고, 주변에 차원테이블(Dimensional Table)을 둔다.
    - 펙트테이블: 각 차원과 관련된 모든활동이 기록되는 테이블이다. 
        - ex_판매액
    - 차원테이블: 주로 데이터를 분류하기 위한 속성들이 저장되어 있다. 

- 정규화된 스키마를 갖는 형태로 데이터를 저장하고 있는 데이터베이스와 스타 스키마로 저장하고 있는 데이터 웨어하우스는 스키마 구조가 다름.
- 서로 다른 스키마 구조를 가지기때문에, 기존 db를 스타스키마에 맞게 변환을해서 변환데이터를 적재해야함(= ETL이라고한다)

### ETL

- Extract: 원본데이터나 데이터소스로부터 데이터를 가져오는 것
- transform: 데이터 구조를 변경하는 프로세스. 필요에 따라 필터링, 가공을 하며 필요한 형태로 변형. 데이터웨어하우스에 데이터를 넣을 수 있는 형태로 만듦
- Load: 적재에서는 변환된 데이터를 최종 목적지에 저장하는 프로세스를 이야기한다. 앞에서는 데이터 웨어하우스에 저장하는 것을 예로 들었지만, 데이터베이스나 다른 저장소에도 저장할 수 있다.

** 최근에는 ELT라는 패턴도 많이 사용되고 있다.
- ETL은 데이터를 추출하고 변환 후 적재하는 것이었다면, ELT는 데이터를 추출하고 적재해 변환하는 것이다.

### ELT

- ELT가 생긴 이유는 데이터가 커지고 다양해지면서 ETL과정이 굉장히 복잡해짐
    - 이미지, 동영상, 음성파일 등은 데이터 웨어하우스에 넣을 수 없는 데이터이다.
- 그래서 데이터를 다 적재한 후에 어떻게 처리할지 결정한다.
- ELT는 데이터 웨어하우스에도 적재하지만, 데이터 레이크(Data Lake)에도 적재한다.

**데이터 레이크는 무엇일까?

## 3-2. 데이터 레이크(Data Lake)

## 3-2-1. 데이터 레이크 정의 및 특징
데이터 마트를 일종의 가게. 즉, 물이 정화되고, 포장되고, 가게에 진열되면서 소비하기 쉽게 된 상태라고 생각해보자. 데이터 레이크는 좀 더 물이 자연상태에 가까운 큰 물줄기라고 볼 수 있다.

**데이터들이 soruce로 부터 흘러와서 Data Lake를 채우고, 여러 사용자가 Lake에서 데이터를 확인하거나 필요에따라 표본을 가져와 처리할 수 있는 것

- 데이터 레이크는
    - 데이터의 원래 형태와 포멧을 유지해서 저장
    - 대량의 데이터를 저장하고 처리하고 보호하기 위한 저장소이다.
    - 데이터의 종류와 상관없이 저장이 가능함.
    - HDFS, 아마존 S3와 같은 분산 오브젝트 storage에 많이 저장되고 있다.
        - HDFS: Hadoop Distributed File System
        - 아마존 S3: 아마존 S3는 아마존 웹 서비스에서 제공하는 온라인 스토리지 웹 서비스
    - Data Lake에 저장된 데이터는 다양한 사용자들이 접근해서 사용할 수 잇다.

## 3-2-2. 데이터 레이크의 데이터를 계층화
- 1단계: 원본 데이터
- 2단계: 분석용 데이터
- 3단계: 피처 데이터(ML, AI) / 집계 데이터

- 원본 데이터를 그대로 저장한 것
- 원본 데이터로 분석용 데이터를 생성.
    - 분산된 데이터를 빠르게 분석을 하기 위해서 파케이나 orc와 같은 컬럼기반으로 저장.
        - 파케이(parquet): 하둡에서 칼럼방식으로 저장하는 저장 포맷을 말합니다.
        - ORC는 Hive에 최적화된 형식이고, Parquet는 스파크에 최적화된 형식입니다.
- 일반적인 분석데이터가 아닌, ML이나 AI에서 활용할 데이터를 저장하게 된다, 집계 데이터 역시 3단계 데이터임

**계층별로 구분하여 나누어 볼 수 있다.

# 4. Data Lake vs Data Warehouse

1. 데이터 레이크
    - 원시데이터
    - 일단 데이터를 저장하고 사용해서 읽어들일때 스키마가 적용되어서 데이터를 읽음. 읽기스키마라고함
    
    
2. 데이터 웨어하우스
    - 데이터를 구조화된 모델로 변경해서 저장
    - 데이터를 적재하기 전에 스키마에 맞게 데이터를 변경하여야한다. 스키마에 어떤 형태로 저장될지 미리 정함. 쓰기스키마라고함
    

### 최근 데이터파이프라인의 패턴 트랜드는 ETL -> ELT로 바뀌는 추세이다.
- 왜? 더 유연하니까. 데이터가 워낙 크고 다양해지다보니까 일단 저장해놓고 필요한걸 사용할 때 스키마 적용. 

> ELT로 구성하게되면 데이터엔지니어는 추출과 저장에 집중할 수 있고, 데이터분석가 데이터사이언티스트는 변환에 집중할 수 있어서 역할을 명확하게 나눌 수 있음.

# 5. 데이터 파이프라인 오케스트레이션
- 데이터 파이프라인 워크플로우를 관리해주는 오케스트레이션 종류를 살펴보고, 각 프레임워크를 비교해서 살펴보겠다.

### 5-1. 오케스트레이션이란?
- 여러 개의 컴퓨터 시스템, 애플리케이션 또는 서비스를 조율하고 관리하는 것
- 복잡한 태스크와 워크플로를 쉽게 관리할 수 있도록 도와줌

**데이터 파이프라인을 구축해두면 하나의 job만 수행하는 것이 아니라 다양한 job들을 수행해서 여러 데이터를 가공하고 저장할 수 있다.

- 여러 task들을 실행하면서 데이터를 처리하는 것을 workflow라고 한다.

### 5-2. workflow관리 도구의 필요성
- 스크립트의 한계
    - workflow가 복잡해짐
    - task의 의존 관계가 복잡 -> 스크립트를 통해서 하긴 어렵다.
    - 실패 시 처리가 어렵다. -> 복잡한 의존관계를 설정하게 되었을 때, 실패하면 연관되어있으니 처음부터 다시 실행해야함 

**이러한이유로 workflow관리 도구가 필요

### 5-3. workflow관리 도구 기능
- 스케줄링 -> 크론탭으로 더 편하게 가능
    - 크론탭(crontab): 서버 스스로 정해둔 명령을 일정 기간마다 실행하도록 하는 것입니다
- 태스크 의존 관계 정의 -> task들이 빠짐없이 실행되는 것을 보장해줌.
- 실행 결과 알림 및 보관 -> task들이 다 돌면 실행 결과 알려주고 통제해줌
- 실패 시 재실행 -> 실패한 테스크부터 재실해주는 기능 제공

**workflow 관리도구는 데이터 파이프라인의 모든 태스크들의 관리를 손쉽게 해주는 다양한 기능들을 제공해준다.

### 5-4. DAG(Directed Acyclic Graph) [ 댁,덱 이라고 읽음]
- workflow내 task들의 의존관계는 어떻게 표현할 수 있을까?

- 방향성(Directed)
    - 간선에 방향이 존재
- 비순환(Acyclic)
    - 그래프의 한 정점에서 시작하여 다시 시작 정점으로 이어지는 간선이 존재하지 않음
    - 사이클이 존재하지 않음
    

#### DAG의 예시로 게임 스킬트리가 있다.
- 순서에 맞춰서 찍어줘야함.

### 실제 task들의 의존관계를 표현한 그림
- DAG의 형태로 표현 되어있다(화살표, acyclic)
    - 이러한 이유로 task의 의존방향을 DAG의 형태로 표현한다.

## 5-5. Airflow / oozie에서 DAG을 표현하는 방식
- Airflow[에어플로우]
- oozie[우지]

## 5-5-1. Airflow
- 4개의 task를 정의하고 
- task의 의존관계를 표현해줌
- task1 다음에 task2,task3이 실행되고 그 후 task4 실행
- 파이썬으로 추가적인 처리 불가

## 5-6. 에어플로우와 우지의 아키텍쳐

- DB내에 task들의 실행결과를 저장한다.
- workflow관리 도구들은 실행이 종료되면 상태에 따라 알림을 보내줌. 그 알림보고 대응 가능. 
- 실행정보를 가지고 있기 떄문에 언제 실행되고 언제 종료되었는지 파악가능하고 여러 장점이있음
- 실패 시 복구 방법
    - 재시도
    - backfill

## 5-7. 실패 시 재실행 할 수 있는 기능을 제공한다
- 재시도
- backfill

### 재시도
- 재시도 설정을 해둠. 일시적인 문제의 경우, 자동으로 재시작하여 성공적으로 마무리 되게끔 해줌. 
- 재시도 횟수가 너무 많으면, 태스크의 오류로 인한 실패 상황을 늦게 발견해서 대응이 늦을 수 있음
- 재시도 횟수가 너무 적으면, 일시적문제가 해결되기전에 실패가 될 수도 있다. 

**태스크의 특성을 잘 이해하고 적절히 재시도를 설정하자. 2~3회를 추천. 

### backfill
- 일정 기간동안 workflow를 재실행하는 기능
- 며칠동안 실패하는 경우, 모아서 재실행.
- 혹은 새로운 workflow를 만들고, 과거부터 실행하고 싶을때 사용.
    - ex_한 달 전 데이터부터 처리 한다고 할 때, 하루에 처리하는 양이 아닌 한달치를 작업하는 것이므로 부하 주의.
