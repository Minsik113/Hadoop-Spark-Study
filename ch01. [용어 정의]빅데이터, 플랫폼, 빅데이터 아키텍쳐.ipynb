# 1 빅데이터의 정의
- 기존의 데이터 처리 응용 소프트웨어로는 수집·저장·분석·처리하기 어려울 정도로 방대한 양의 데이터
- 데이터로부터 가치를 추출하고 결과를 분석하는 기술

- 3V: volumn(데이터양), velocity(속도), variety(다양성)
    - 데이터양: TB, PB이상의 굉장히 큰 데이터 양임
    - 속도: 오늘날 데이터가 생성되는 시점부터 데이터를 처리 결과가 사용자에게 전달되는 시점까지의 시간이 점점 짧아지고 있음. 적시에 맡게 빠르게 수집, 저장, 분석이 이루어져야한다. 
    - 다양성: 다양한 소스로부터 다양한 형식의 데이터를 처리할 수 있어야 한다. 구조화된 데이터 or 반구조화 데이터들이 굉장히 많고 이를 저장하고 처리할 수 있어야함
    
- 이러한 이유로 빅데이터는 데이터 크기만이 아닌, 다양한 유형의 데이터로부터 가치를 찾아내는 것을 의미.

# 1-1 빅데이터로부터 기업들이 얻을 수 있는 이유
1. 빠르고 더 나은 의사 결정 지원
2. 새로운 전략과 제품 구축
3. 고객에 관한 통찰력 향상
4. 가까운 미래를 예측하고 새로운 기회 창출

- 페이스북: 소셜 네트워킹 분석 및 광고 최적화
- 아마존: 고객 제품 추천
- 넷플릭스: 컨텐츠 추천
- Uber: 예측을 통한 배차 시스템 최적화

> 이러한 이유로 빅데이터는 선택이 아닌 필수이다.

- 대부분의 기업이 데이터 분석과 예측을 위한 플랫폼 구축 필요
    - garbage in, garbage out : 불필요한 데이터는 빠르게 버려야함.
- 데이터 분석과 예측을 위한 기반을 마련하는데 필수
- 데이터를 잘 모으고, 관리해야 함

# 2. 플랫폼이란
- 많은 사람들이 쉽게 이용할 수 있고 다양한 목적의 비즈니스가 이루어지는 공간

# 2-1. 하드웨어 플랫폼
- 동일한 제품을 일정한 품질로 만드는 프로세스와그 제품을 만드는 장치


# 2-2. 소프트웨어 플랫폼
- 소프트웨어를 실행할 수 있는 기반이나 환경


- os에 따라 맞춤 소프트웨어를 만들지. 하지만 java로 jvm위에 올려 해결
- 그래서 자바 애플리케이션 개발자는 os나 하드웨어에 독립적인 애플리케이션을 개발할 수 있는 것임

# 2-3. 서비스 플랫폼
- 서비스 제공자의 서비스를 다른 서비스들이 쉽게 사용할 수 있게 해주는 환경

- 물건 사고팔 수 있는 쇼핑 플랫폼
- 앱스토어
    - 앱 판매자와 구매자들이 있음

# 2-4. 빅데이터 플랫폼
- 방대한 양의 데이터로부터 가치를 추출하여 결과를 분석
- 플랫폼은 많은 사용자가 이용할 수 있는 공간

- 즉,기업 내에 많은 사용자들이 데이터를 처리하고 분석을 쉽게 할 수 있는 환경을 제공해주는 시스템
- 데이터들의 원천이 다 다름. 어떻게 처리하고 저장할 것인지, 어디에 저장할지, 어떻게 분석할지, 어떻게 보여줄지.

1. 데이터 수집, 처리 및 저장
2. 데이터 발견, 검색, 보안 제공
3. 데이터 분석 및 ML 지원 -> 예측할 수 있는 환경 제공해 줘야함

# 2.5 빅데이터 아키텍쳐

- 프로젝트 유형마다 다양함 -> 다양한 아키텍쳐로 구성됨
- 공부하면서 아키텍쳐의 이해도가 올리자. 빅데이터 플랫폼을 구축하기 위한 통찰력을 공부하자
