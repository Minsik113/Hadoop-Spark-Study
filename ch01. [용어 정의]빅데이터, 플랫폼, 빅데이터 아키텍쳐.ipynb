# 이 글 보면 빅데이터 아키텍쳐의 청사진이 그려짐

# 1 빅데이터의 정의
- 기존의 데이터 처리 응용 소프트웨어로는 수집·저장·분석·처리하기 어려울 정도로 방대한 양의 데이터
- 데이터로부터 가치를 추출하고 결과를 분석하는 기술

- 3V: volumn(데이터양), velocity(속도), variety(다양성)
    - 데이터양: TB, PB이상의 굉장히 큰 데이터 양임
    - 속도: 오늘날 데이터가 생성되는 시점부터 데이터를 처리 결과가 사용자에게 전달되는 시점까지의 시간이 점점 짧아지고 있음. 적시에 맡게 빠르게 수집, 저장, 분석이 이루어져야한다. 
    - 다양성: 다양한 소스로부터 다양한 형식의 데이터를 처리할 수 있어야 한다. 구조화된 데이터 or 반구조화 데이터들이 굉장히 많고 이를 저장하고 처리할 수 있어야함
    
- 이러한 이유로 빅데이터는 데이터 크기만이 아닌, 다양한 유형의 데이터로부터 가치를 찾아내는 것을 의미.

# 1-1 빅데이터로부터 기업들이 얻을 수 있는 이유
1. 빠르고 더 나은 의사 결정 지원
2. 새로운 전략과 제품 구축
3. 고객에 관한 통찰력 향상
4. 가까운 미래를 예측하고 새로운 기회 창출

- 페이스북: 소셜 네트워킹 분석 및 광고 최적화
- 아마존: 고객 제품 추천
- 넷플릭스: 컨텐츠 추천
- Uber: 예측을 통한 배차 시스템 최적화

> 이러한 이유로 빅데이터는 선택이 아닌 필수이다.

- 대부분의 기업이 데이터 분석과 예측을 위한 플랫폼 구축 필요
    - garbage in, garbage out : 불필요한 데이터는 빠르게 버려야함.
- 데이터 분석과 예측을 위한 기반을 마련하는데 필수
- 데이터를 잘 모으고, 관리해야 함

## a. 데이터 베이스

### OLTP vs OLAP
- OLTP
    - Online Transaction Processing
    - ex_회원정보 수정, 상품주문, 댓글 남기기 및 수정
    - 무수히 많이 발생되는 각각의 작업요청을 오류없이 처리하고, 그 결과값을 실시간으로 확인시켜줘야 함을 의미한다. 즉, 1개의 트랜잭션(atm기에 카드넣고 돈이 제대로 전달되었는지의 일련과정)에서 발생되는 insert, update, delete의 과정을 무결성을 보장하여 처리하고 그 결과를 select 하는 과정을 OLTP라고 부른다.
- OLAP
    - Online Analytical Processing
    - ex_회사 db에 정보로 1년치 사용내역을 조회
    - 기존에 저장되어 있는 데이터를 사용자의 요구와 목적에 맞게 분석하여 정보를 제공하는 개념을 OLAP라고 한다.


### 이벤트 컬렉터
- db의 데이터는 비즈니스 사이드에서 생성되는 데이터라면, 이벤트 컬렉터는 사용자가 만드는 데이터이다.

### 로그
- 일반적으로 서버의 로그라고 의미한다.
- 서버들의 로그를 모아서 분석을 하거나 로그로 활용

### API
- 필요에 의해 외부 서비스를 사용하게 될 수 있는데, 이러한 외부서비스를 가져오기위해 해당 서버로부터 데이터를 가져오기 위해 API이용
- ex_신용정보조회

# 3-2. Data Ingetsion and Processing 단계
- 데이터를 수집하고 처리하는 과정
- ETL이라는 추출 변형이 이루어지는 과정
    - E(Extract), T(Transform), L(Load)
    - 기업이 전 세계 모든 곳의 수많은 팀에서 관리하는 구조화된 데이터와 구조화되지 않은 데이터를 비롯한 전체 데이터를 가져와 비즈니스 목적에 실질적으로 유용한 상태로 변환하는 end-to-end 프로세스를 의미한다.
    - 조직에서 여러 시스템의 데이터를 단일 데이터베이스, 데이터 저장소, 데이터 웨어하우스 또는 데이터 레이크에 결합하기 위해 일반적으로 허용되는 방법.

## 스트림처리 단점
- 잘못된 결과를 처리하기 어렵다.
- 지연된 데이터 처리에 대한 문제.
    - 집계가 완료되었는데 이후에 데이터가 들어온 경우 문제가 생김

## 실시간 처리를 위한 대표적인 두 가지 아키텍쳐
- labmda 아키텍쳐 
- kappa 아키텍쳐

1. 람다(Lambda) 아키텍처
    1. 정의: 동일한 시스템을 처리하는 배치시스템과 스트리밍 시스템을 함께 사용하는 것
    2. 3개의 레이어로 구성(batch 레이어, serving 레이어, speed 레이어)
    3. 데이터가 스피드 레이어로 들어와 처리가 된다.
    4. 처리 된 데이터는 real-time view에 업데이트가 되어 결과가 보여진다.
    5. 배치 레이어에서는 과거 데이터가 들어와 처리된다.
    6. 이러한 배치 레이어에서는 집계를 여러번이고 다시 할 수 있다.
    7. 대규모의 배치처리를 하기 때문에 실행 시간이 좀 오래 걸리는 특성이 있다.
    8. 배치 처리 결과는 batch view로 이동한 후 serving layer에서 처리가 된다.
    9. batch views가 완료된 시점에서, real-time views가 삭제가 된다. 리얼타임 뷰는 배치가 업데이트전까지만 유지된다.
    10. 배치뷰와 실시간뷰를 조합하여 쿼리를 시행하게 된다. merged view라는 곳에서 쿼리로 실행하게 된다
- 즉, 24시간이내는 실시간뷰에서 참조하고, 그 이후에 대한 정보는 배치뷰로 확인한다는 이야기이다.
- 단점은 효율적이지 않다.
    - 똑같은 처리(실시간처리와 배치처리)가 2개가 구현되어 관리된다.
    - 운영에 있어서 번거로움이 존재한다.

> 이러한 단점을 개선하기 위해 카파(Kappa) 아키텍처가 등장했다.

2. 카파(Kappa) 아키텍처
- 스트리밍 시스템에서도 강력한 일관성을 제공하기 시작함
- 람다 아키텍처와 비교해보면
    - batch layer가 없다. 보다 단순화된 형태임
    - 메세지 브로커가 다양한 기능을 제공해줘야함.
        - 데이터 보관을 충분히 하면서 문제가 발생하였을 때, 문제 발생시점으로 되돌아가서 처리를 해야함.
        - 이러한 방식으로 과거의 문제를 처리할 수 있음

- 스트림의 처리 결과가 멱등성을 지녀야함. 
    - 여러번 처리되어도 동일한 결과가 나올 수 있어야함
- 스트림처리만으로 구성할 수 있는 아키텍처가 카파 아키텍처이다.
- 대량의 데이터를 보낼 때 평상시보다 부하가 높아짐.
    - 최근에 클라우드 서비스에서 스케일링 서비스를 잘 제공하기 때문에 잘 처리할 수 있다.
    

# 3-3. Storage 단계
- 앞단계에서 데이터를 처리한 것을 저장하는 단계이다.
- 주로 data warehouse, data lake에 저장하고 NoSQL에 저장도 가능
- 쿼리를 하거나 또 다른 데이터 처리가 가능한 형태로 데이터를 저장한다.
- 데이터 레이크의 경우, Raw한 형태로 저장한다.
    - Raw한 형태: 데이터에 다른 처리를 하지 않고 원천 그대로 저장한다는 의미

## 3-3-1. Data Warehouse

- 서로 다른 시스템에 데이터가 모델링이 되어있는 시스템
- 정형화된 데이터가 들어가 있다.
    - 리포터나 분석에 최적화 되어있음
- 최근에는 온프레임위스(?)기반의 데이터 웨어하우스보다 '클라우드 기반의 데이터 웨어하우스'를 사용하는 추세이다. 위에 보이는거.
- 일반적인 RDBMS는 TB크기의 데이터를 지원한다. TB보다 커지면 성능저하 발생
- 최신 data warehouse들은 수 PB를 지원한다.
1. 구글의 BigQuery(빅쿼리)
    - 구글 cloud platform에서 중요한 역할을 함.
    - 여기에서 쿼리를 실행한다. 
    - 데이터를 복제해서 저장하므로 안정성이 뛰어나고 배치와 스트리밍으로 전달 가능하다.
2. 아마존의 redshift(레드시프트)
    - managed형 cloud data warehouse 서비스이다.
    - 열기반 방식으로 데이터를 압축하여 분산저장함.
    - 컬럼기반으로 저장함.
    - 테이블 간 효율을 개선하기위해 병렬처리
3. snowflake(스노우 플레이크)
    - 싸스. 소프트에저서비스(?)라는 모델
    - 정형/ 비정형 데이터 모두를 지원하는 데이터웨어하우스를 지원한다
    - 자체적인 클라우드 서비스가 없기때문에 aws나 gcp에서 구동됨.

## 3-3-2. Data Lake

- 다양하고 방대한 유형의 데이터 저장 가능
- 아마존의 s3, hdfs, 다른 클라우드 오브젝트 storage를 사용한다.
- 최근에는 Data Lake들이 발전하기 시작.
    - 기존의 저장소 위에서 또다른 storage를 구축해서 한단계 진화된 형태.
    - storage layer를 제공하는 프로젝트들이 apach푸디(?)나 델타레이크(?) 아이스버그(?)들이 있다.

#### 기존의 data lake 문제점
- 트랜젝션을 지원하지 못하거나 데이터의 업데이트나 삭제가 불가능했다.
- 스토리지 레이어들은 이러한 기능들을 제공해줌으로써 Data Lake가 한단계 발전하고있다.
- 최근에는 Data Lake House라고 부름

# 3-5. Output 단계
- 데이터 분석 결과를 보여주고, 애플리케이션에 실제로 적용하는 단계이다.
- 분석결과는 대시보드와 같은 시각화 도구를 사용

### 3-5-1. 데이터 플랫폼을 구축하는 목적
- 처리된 데이터 결과로 의사결정을 할 수 있게 도와준다.
    - 비즈니스 의사 결정을 도와줌
- 데이터 기반의 머신러닝 딥러닝 모델을 사용하여 서비스 질을 높인다.
    - 제품의 질을 높임

### 3-5-2. 대표적인 시각화 도구


1. Tableau [태블로]
    - 가장 유명함
    - 데이터 시각화 기능 제공. 시각화한 기능으로 분석도 가능. 시각화 된 정보를 대시보드에 보여줄 수 있음
2. Apache Superset[아파치 슈퍼셋]
    - 태블로와 달리 오픈소스이다.
    - 대시보드생성
    - 공짜라 태블로에 비해 보안 약함
3. redash [리대시]
    - 오픈소스
    - 대시보드생성.
    - 스케줄링 제공

# 3-6. Workflow Management 워크플로우 관리도구
- 앞서 말한 1~5단계들을 연결하여 파이프라인을 구성한다.
- 파이프라인 작업을 할 때에는 파이프라인에 대한 오케스트레이션 및 스케줄링, 상호 의존관계를 구성하는 workflow관리가 필요
- 이를 지원해주는게 workflow관리 도구
    - 정기적으로 task실행하여 비정상적인 상태를 감지하여 정상적인 상태로 되돌리게 지원해준다.
